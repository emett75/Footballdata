```python
plt.plot(nfl['WPA'], color='purple')
plt.plot(nfl['Win_Prob'], color='green')
plt.plot
```

```python
#according to the above graph, there is a correlation between win probability and WPA, or win probability added by individual players.
#this data shows us that when single players are able to positively influence a game, the probability of a win goes up.
#also, it shows us that players can negatively influence the outcome as well.
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(nfl['WPA'], color = 'purple')
plt.ylabel('WPA')
plt.title('Win Prob Added')

plt.subplot(1, 2, 2)
plt.plot(nfl['Win_Prob'], color = 'green')
plt.ylabel('Win_Prob')
plt.title('Win Probability')
plt.show
```

```python
#using the subplot function, I took a closer look at the same data as the previous table. 
#using this subplot, we also see that there is a correlation between the WPA and the overall win probability.
#Using this data, we can see that a single player can change the outcome of a game.
```

```python
plt.scatter(
    x = nfl['WPA'], 
    y = nfl['airWPA'],
    color='green'
)

plt.scatter(
    x = nfl['WPA'],
    y = nfl['yacWPA'],
    color=['red']
)
plt.show()
```

```python
#this graph shows the correlation between WPA and airWPA versus WPA and yacWPA. In most cases, they are very equal, with a few outliers.
```
